## 6. ParÃ¢metros de Controle

Investigar parÃ¢metros disponÃ­veis para ajuste fino.

```python
# PARÃ‚METROS ESPERADOS (baseado em papers de diffusion TTS)

"""
PossÃ­veis parÃ¢metros F5-TTS:

1. **nfe_step** (Number of Function Evaluations)
   - Similar a "steps" em modelos de difusÃ£o
   - Range: 10-50 (mais = melhor qualidade, mais lento)
   - Default: ~32

2. **cfg_strength** (Classifier-Free Guidance)
   - Controla aderÃªncia ao texto
   - Range: 1.0-3.0
   - Default: ~2.0

3. **speed** (Velocidade de fala)
   - Range: 0.5-2.0
   - Default: 1.0

4. **sway_sampling_coef**
   - Controle de variaÃ§Ã£o/naturalidade
   - Range: 0.0-1.0
   - Default: -1 (auto)

COMPARAÃ‡ÃƒO XTTS vs F5-TTS:

XTTS:                    F5-TTS (esperado):
- temperature            - nfe_step
- top_p                  - cfg_strength
- top_k                  - sway_sampling_coef
- repetition_penalty     - (nÃ£o aplicÃ¡vel)
- speed                  - speed
"""

print("ğŸ“Š ParÃ¢metros F5-TTS (baseado em literatura):")
print("   - nfe_step: Qualidade (10-50, default ~32)")
print("   - cfg_strength: AderÃªncia ao texto (1.0-3.0)")
print("   - speed: Velocidade de fala (0.5-2.0)")
print("   - sway_sampling_coef: Naturalidade (0.0-1.0)")
```
